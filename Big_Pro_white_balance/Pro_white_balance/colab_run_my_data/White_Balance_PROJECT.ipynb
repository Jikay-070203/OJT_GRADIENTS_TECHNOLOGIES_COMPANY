##Test 1

```python
# Thư viện cần thiết
import PIL
import torch
from PIL import ImageOps
from diffusers import StableDiffusionInstructPix2PixPipeline, EulerAncestralDiscreteScheduler
import matplotlib.pyplot as plt

# Load mô hình Instruct-Pix2Pix
model_id = "timbrooks/instruct-pix2pix"
pipe = StableDiffusionInstructPix2PixPipeline.from_pretrained(model_id, torch_dtype=torch.float16, safety_checker=None)
pipe.to("cuda")
pipe.scheduler = EulerAncestralDiscreteScheduler.from_config(pipe.scheduler.config)

# Đọc ảnh từ Google Drive
input_image_path = '/content/drive/MyDrive/OJT/images/IMG_1683.png' #thay path
image = PIL.Image.open(input_image_path)

# Định dạng RGB
image = ImageOps.exif_transpose(image)
image = image.convert("RGB")

# Câu lệnh prompt
prompt = "photo white balance"

# Tiến hành tạo ảnh output với mô hình
images = pipe(prompt, image=image, num_inference_steps=10, image_guidance_scale=1).images

# Lưu ảnh output
output_image_path = '/content/drive/MyDrive/OJT/images/result_colab/output_1683.jpg'  # Thay đổi path
images[0].save(output_image_path)  # Lưu ảnh output
# Hiển thị ảnh input và output
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Hiển thị ảnh input bên trái
axes[0].imshow(image)
axes[0].set_title('Input Image')
axes[0].axis('off')  # Ẩn trục

# Hiển thị ảnh output bên phải
axes[1].imshow(images[0])
axes[1].set_title('Output Image')
axes[1].axis('off')  # Ẩn trục

# Hiển thị kết quả
plt.show()
```
