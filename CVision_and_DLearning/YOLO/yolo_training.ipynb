# TEST MODEL ONNX

```python
import onnxruntime
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt

# ⚙️ Cấu hình
image_path = '/content/drive/MyDrive/SourceCode/OJT/Yolo_training/test/face_test.jpg'
onnx_model_path = '/content/drive/MyDrive/SourceCode/OJT/Yolo_training/model/Pytorch_model/best.onnx'
CONF_THRESHOLD = 0.4
IOU_THRESHOLD = 0.5
CLASSES = ['face']  # 🛠️ Sửa theo nhãn của bạn
COLORS = [(0, 255, 0)]

# 📤 Hàm chuyển bbox
def xywh2xyxy(x):
    y = np.copy(x)
    y[0] = x[0] - x[2] / 2
    y[1] = x[1] - x[3] / 2
    y[2] = x[0] + x[2] / 2
    y[3] = x[1] + x[3] / 2
    return y

# 🧠 Load model ONNX
session = onnxruntime.InferenceSession(
    onnx_model_path,
    providers=['CUDAExecutionProvider', 'CPUExecutionProvider']
)

# 🖼️ Load & chuẩn hóa ảnh
img = cv2.imread(image_path)
orig_img = img.copy()
img_resized = cv2.resize(img, (640, 640))
img_input = img_resized[..., ::-1].transpose(2, 0, 1)  # BGR → RGB → CHW
img_input = np.expand_dims(img_input, axis=0).astype(np.float32) / 255.0

# ▶️ Inference
input_name = session.get_inputs()[0].name
outputs = session.run(None, {input_name: img_input})
pred = np.squeeze(outputs[0]).T  # (8400, 5)

# 📦 Post-processing
boxes = []
scores = []
class_ids = []

for i in range(pred.shape[0]):
    x, y, w, h, conf = pred[i][:5]
    if conf > CONF_THRESHOLD:
        box = xywh2xyxy([
            x * orig_img.shape[1],
            y * orig_img.shape[0],
            w * orig_img.shape[1],
            h * orig_img.shape[0]
        ])
        boxes.append([int(b) for b in box])
        scores.append(float(conf))
        class_ids.append(0)  # Chỉ có 1 class

# 🧹 NMS
indices = cv2.dnn.NMSBoxes(boxes, scores, CONF_THRESHOLD, IOU_THRESHOLD)

# 🎨 Vẽ kết quả
if len(indices) > 0:
    for i in indices.flatten():
        x1, y1, x2, y2 = boxes[i]
        conf = scores[i]
        cls_id = class_ids[i]
        label = f"{CLASSES[cls_id]}: {conf:.2f}"
        color = COLORS[cls_id % len(COLORS)]

        cv2.rectangle(orig_img, (x1, y1), (x2, y2), color, 2)
        cv2.putText(orig_img, label, (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

# 📷 Hiển thị kết quả
print("Output shape:", outputs[0].shape)
plt.imshow(cv2.cvtColor(orig_img, cv2.COLOR_BGR2RGB))
plt.title("Kết quả từ ONNX")
plt.axis(False)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from PIL import Image

# 🔧 Cấu hình
image_path = '/content/drive/MyDrive/SourceCode/OJT/Yolo_training/test/test1.jpeg'
tflite_model_path = '/content/drive/MyDrive/SourceCode/OJT/Yolo_training/model/TensorFlow_model/best_float32.tflite'
CONF_THRESHOLD = 0.4
IOU_THRESHOLD = 0.5
CLASSES = ['face']
COLORS = [(255, 0, 0)]

# 📤 Hàm chuyển bbox
def xywh2xyxy(x):
    y = np.copy(x)
    y[0] = x[0] - x[2] / 2
    y[1] = x[1] - x[3] / 2
    y[2] = x[0] + x[2] / 2
    y[3] = x[1] + x[3] / 2
    return y

# 🧠 Load model TFLite
interpreter = tf.lite.Interpreter(model_path=tflite_model_path)
interpreter.allocate_tensors()
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# 📷 Load ảnh
img = cv2.imread(image_path)
orig_img = img.copy()
img_resized = cv2.resize(img, (640, 640))
img_input = img_resized[..., ::-1].astype(np.float32) / 255.0  # BGR → RGB + normalize
img_input = np.expand_dims(img_input, axis=0)

# ▶️ Inference
interpreter.set_tensor(input_details[0]['index'], img_input)
interpreter.invoke()
output_data = interpreter.get_tensor(output_details[0]['index'])  # (1, 8400, 5)
pred = np.squeeze(output_data).T  # (5, 8400) → (8400, 5)

# 📦 Post-processing
boxes, scores, class_ids = [], [], []
for i in range(pred.shape[0]):
    x, y, w, h, conf = pred[i][:5]
    if conf > CONF_THRESHOLD:
        box = xywh2xyxy([
            x * orig_img.shape[1],
            y * orig_img.shape[0],
            w * orig_img.shape[1],
            h * orig_img.shape[0]
        ])
        boxes.append([int(b) for b in box])
        scores.append(float(conf))
        class_ids.append(0)

indices = cv2.dnn.NMSBoxes(boxes, scores, CONF_THRESHOLD, IOU_THRESHOLD)

# 🎨 Vẽ kết quả
for i in indices.flatten():
    x1, y1, x2, y2 = boxes[i]
    conf = scores[i]
    cls_id = class_ids[i]
    label = f"{CLASSES[cls_id]}: {conf:.2f}"
    color = COLORS[cls_id % len(COLORS)]
    cv2.rectangle(orig_img, (x1, y1), (x2, y2), color, 2)
    cv2.putText(orig_img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

# 🖼️ Hiển thị
print("TFLite output shape:", output_data.shape)
plt.imshow(cv2.cvtColor(orig_img, cv2.COLOR_BGR2RGB))
plt.title("Kết quả từ TFLite")
plt.axis(False)
plt.show()
```

# TEST Model OpenVINO

```python
from openvino.runtime import Core
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image

# ---------- Hàm Non-Maximum Suppression ----------
def non_max_suppression(boxes, scores, iou_threshold=0.5):
    boxes = np.array(boxes)
    scores = np.array(scores)
    indices = cv2.dnn.NMSBoxes(boxes.tolist(), scores.tolist(), score_threshold=0.25, nms_threshold=iou_threshold)
    return indices.flatten() if len(indices) > 0 else []

# ---------- Load model OpenVINO ----------
core = Core()
model_path = "/content/drive/MyDrive/SourceCode/OJT/Yolo_training/model/Openvino_model/best.xml"
compiled_model = core.compile_model(model_path, "CPU")

# ---------- Load ảnh ----------
image_path = '/content/drive/MyDrive/SourceCode/OJT/Yolo_training/test/test2.jpg'
original_img = cv2.imread(image_path)
img_resized = cv2.resize(original_img, (640, 640))
img_input = img_resized[..., ::-1].transpose(2, 0, 1)
img_input = np.expand_dims(img_input, axis=0).astype(np.float32) / 255.0

# ---------- Inference ----------
input_layer = compiled_model.input(0)
output_layer = compiled_model.output(0)
result = compiled_model([img_input])[output_layer]

# ---------- Postprocess ----------
conf_threshold = 0.3
boxes = []
scores = []

for det in result[0].T:  # (5, 8400)
    if det[4] > conf_threshold:
        x_center, y_center, w, h = det[0:4]
        x1 = int((x_center - w / 2) * original_img.shape[1] / 640)
        y1 = int((y_center - h / 2) * original_img.shape[0] / 640)
        x2 = int((x_center + w / 2) * original_img.shape[1] / 640)
        y2 = int((y_center + h / 2) * original_img.shape[0] / 640)
        boxes.append([x1, y1, x2 - x1, y2 - y1])  # Format cho NMS: [x, y, w, h]
        scores.append(float(det[4]))

# ---------- Apply NMS ----------
indices = non_max_suppression(boxes, scores, iou_threshold=0.5)

# ---------- Vẽ kết quả ----------
for i in indices:
    x, y, w, h = boxes[i]
    conf = scores[i]
    label = f'Face {conf:.2f}'

    # Vẽ khung với độ dày lớn hơn (thickness = 6)
    cv2.rectangle(original_img, (x, y), (x + w, y + h), (0, 255, 0), 6)

    # Hiển thị nhãn với font lớn hơn và nét chữ dày (thickness = 3)
    cv2.putText(original_img, label, (x, y - 15),
                cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 255, 0), 3)

# ---------- Hiển thị ----------
print("Output shape:", result.shape)
plt.imshow(cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB))
plt.title("Kết quả từ OpenVINO")
plt.axis(False)
plt.show()
```

```python
import torch
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# -------- Load TorchScript model ----------
model_path = '/content/drive/MyDrive/SourceCode/OJT/Yolo_training/model/Pytorch_model/best1.torchscript'
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = torch.jit.load(model_path, map_location=device).eval().to(device)

# -------- Load và xử lý ảnh ----------
image_path = '/content/drive/MyDrive/SourceCode/OJT/Yolo_training/test/test3.jpg'
original_img = cv2.imread(image_path)
img_resized = cv2.resize(original_img, (640, 640))
img = img_resized[..., ::-1].transpose(2, 0, 1)  # BGR -> RGB -> CHW
img = np.expand_dims(img, 0).astype(np.float32) / 255.0
img_tensor = torch.from_numpy(img).to(device)

# -------- Inference ----------
with torch.no_grad():
    preds = model(img_tensor)

# -------- Postprocess ----------
conf_threshold = 0.3
preds_np = preds.cpu().numpy().T  # (8400, 5 + num_classes)
num_classes = preds_np.shape[1] - 4  # 4 bbox + scores

boxes = []
scores = []
class_ids = []

h_orig, w_orig = original_img.shape[:2]

for d in preds_np:
    class_scores = d[4:]
    class_id = np.argmax(class_scores)
    score = class_scores[class_id]

    if score > conf_threshold:
        x_center, y_center, w, h = d[:4]
        x1 = int((x_center - w / 2) * w_orig / 640)
        y1 = int((y_center - h / 2) * h_orig / 640)
        x2 = int((x_center + w / 2) * w_orig / 640)
        y2 = int((y_center + h / 2) * h_orig / 640)
        boxes.append([x1, y1, x2 - x1, y2 - y1])
        scores.append(float(score))
        class_ids.append(int(class_id))

# -------- Apply NMS --------
indices = cv2.dnn.NMSBoxes(boxes, scores, conf_threshold, 0.5)

# -------- Vẽ bounding boxes --------
for i in indices:
    i = i[0] if isinstance(i, (list, np.ndarray)) else i
    x, y, w, h = boxes[i]
    score = scores[i]
    class_id = class_ids[i]
    label = f"Face {class_id} | {score:.2f}"

    cv2.rectangle(original_img, (x, y), (x + w, y + h), (0, 255, 0), 6)  # Nét đậm hơn
    cv2.putText(original_img, label, (x, y - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 255, 0), 3)

# -------- Hiển thị ----------
plt.imshow(cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB))
plt.title("Kết quả từ TorchScript")
plt.axis(False)
plt.show()
```
