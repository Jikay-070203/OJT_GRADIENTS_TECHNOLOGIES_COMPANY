```python
import torch
import onnx
import os
from diffusers import UNet2DConditionModel

# Đường dẫn model gốc & nơi lưu kết quả
model_path = "/content/drive/MyDrive/Model_pix_onnx/model/unet"
save_path = "/content/drive/MyDrive/Model_pix_onnx/result"
os.makedirs(save_path, exist_ok=True)

#  Load mô hình UNet 
unet = UNet2DConditionModel.from_pretrained(model_path, torch_dtype=torch.float32)
unet.to("cpu") 

# Dummy input
dummy_sample = torch.randn(1, 8, 64, 64, dtype=torch.float32)  # Batch 1, 8 channels, 64x64
dummy_timestep = torch.tensor([1], dtype=torch.float32)  
dummy_encoder_hidden_state = torch.randn(1, 77, 768, dtype=torch.float32)  

#  Đường dẫn file xuất ONNX
onnx_path = os.path.join(save_path, "unet.onnx")

#  Xuất sang ONNX 
torch.onnx.export(
    unet,
    (dummy_sample, dummy_timestep, dummy_encoder_hidden_state),
    onnx_path,
    input_names=["sample", "timestep", "encoder_hidden_state"],
    output_names=["output"],
    dynamic_axes={"sample": {0: "batch_size"}, "encoder_hidden_state": {0: "batch_size"}},
    opset_version=14,
    do_constant_folding=True,
    export_params=True,  
    keep_initializers_as_inputs=True 
)

# ✅ Load mô hình ONNX đã xuất
onnx_model = onnx.load(onnx_path)


onnx.save_model(
    onnx_model,
    onnx_path,
    save_as_external_data=True,
    all_tensors_to_one_file=True,  
    location="unet.onnx_data" 
)

print(f"✅ Xuất {onnx_path} thành công")
```
